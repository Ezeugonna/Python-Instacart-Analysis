### 4. Creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
df_ord_prod_merge_grouped.head(20)
```

```python
# Agg data with transform() using the 3 steps process
df_ord_prod_merge_grouped['max_order'] = df_ord_prod_merge_grouped.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Creating flag that assigns a "loyalty" with loc()
df_ord_prod_merge_grouped.loc[df_ord_prod_merge_grouped['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_ord_prod_merge_grouped.loc[(df_ord_prod_merge_grouped['max_order'] <= 40) & (df_ord_prod_merge_grouped['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_ord_prod_merge_grouped.loc[df_ord_prod_merge_grouped['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Print the Frequency of "loyalty_flag"
df_ord_prod_merge_grouped['loyalty_flag'].value_counts(dropna = False)
```

```python
df_ord_prod_merge_grouped.head()
```

### 5. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer).

```python
df_ord_prod_merge_grouped.groupby('loyalty_flag').agg({'prices':['mean', 'max', 'std', 'median']})
```

```python
df_ord_prod_merge_grouped.groupby('loyalty_flag').user_id.nunique()
```

### 6. Create a spending flag for each user based on the average price across all their orders

```python
# To determine the mean price of the product purchased by each user, we used Transform()
df_ord_prod_merge_grouped['mean_product_price'] = df_ord_prod_merge_grouped.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Print head of "user_id", "mean_product_price" and "order_number"
df_ord_prod_merge_grouped[['user_id', 'mean_product_price']].head(30)
```

```python
# Creating spending flag for each user based on avg. price
df_ord_prod_merge_grouped.loc[df_ord_prod_merge_grouped['mean_product_price'] < 10, 'spending_flag'] = 'Low spender'
df_ord_prod_merge_grouped.loc[df_ord_prod_merge_grouped['mean_product_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Frequency Check for spending_flag
df_ord_prod_merge_grouped['spending_flag'].value_counts(dropna = False)
```

```python
df_ord_prod_merge_grouped.groupby('spending_flag').user_id.nunique()
```

```python
# To view relevant column of the updated df
df_ord_prod_merge_grouped[['user_id', 'mean_product_price', 'spending_flag']].head(20)
```

### 7. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.

```python
# To determine the median in 'days_since_prior_order', we used groupby and Transform()
df_ord_prod_merge_grouped['median_days_behaviour_orders'] = df_ord_prod_merge_grouped.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Print head of "user_id", "days_since_prior_order" and "order_number"
df_ord_prod_merge_grouped[['user_id', 'days_since_prior_order']].head(30)
```

```python
# Creating frequency flag with loc()
df_ord_prod_merge_grouped.loc[df_ord_prod_merge_grouped['median_days_behaviour_orders'] > 20, 'order_frequency'] = 'Non-frequent customer'
df_ord_prod_merge_grouped.loc[(df_ord_prod_merge_grouped['median_days_behaviour_orders'] <= 20) & (df_ord_prod_merge_grouped['median_days_behaviour_orders'] > 10), 'order_frequency'] = 'Regular customer'
df_ord_prod_merge_grouped.loc[df_ord_prod_merge_grouped['median_days_behaviour_orders'] <= 10, 'order_frequency'] = 'Frequent customer'
```

```python
# Frequency Check for spending_flag
df_ord_prod_merge_grouped['order_frequency'].value_counts(dropna = False)
```

```python
df_ord_prod_merge_grouped.groupby('order_frequency').user_id.nunique()
```

```python
# To view relevant column of the updated df
df_ord_prod_merge_grouped[['user_id', 'days_since_prior_order', 'median_days_behaviour_orders', 'order_frequency']].head(20)
```

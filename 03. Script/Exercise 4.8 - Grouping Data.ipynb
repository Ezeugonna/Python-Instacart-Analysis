## Grouping Data 

```python
# Grouping df by a given column
df.groupby('product_name')
```

## Aggregating Data with agg()

```python
# Aggregating 'order_number' to list all possible 'department_id'
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# Aggregating without using Agg()
df.groupby('department_id')['order_number'].mean()
```

```python
# repeating the same script with dots instead of []
df.groupby('department_id').order_number.mean()
```

## Aggregating Data with transform()

```python
# Performing multiple aggregation statistics
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
# Agg data with transform() using the 3 steps process
df_ord_prod_merge['max_order'] = df_ord_prod_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

## Deriving Columns with loc()

```python
# Creating flag that assigns a "loyalty" with loc()
df_ord_prod_merge.loc[df_ord_prod_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_ord_prod_merge.loc[(df_ord_prod_merge['max_order'] <= 40) & (df_ord_prod_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_ord_prod_merge.loc[df_ord_prod_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Print the Frequency of "loyalty_flag"
df_ord_prod_merge['loyalty_flag'].value_counts(dropna = False)
```
